```python
#1 ATM PIN Verification 
#Your cousin works at a bank and is testing a simple ATM simulator. 
# A user gets up to 3 attempts to enter the correct PIN. After each wrong attempt, the program should print how many tries are left.
# If they run out of attempts, the card gets locked.
#for loop : 

correct_pin = 1234

correct = False
tries = 3

for i in range(tries):
    entered_pin = int(input("Enter PIN: "))

    if entered_pin == correct_pin:
        print("Access Granted")
        correct = True
        break
    else:
        print("Wrong PIN,", tries - i - 1, "tries left")

if correct == False:
    print("Card Locked")

#correct_pin = 1234 stores the correct ATM PIN.correct = False checks whether the correct PIN is entered.
#tries = 3 sets the maximum number of attempts.
#The for loop runs 3 times, asking the user to enter a PIN each time.
#If the entered PIN matches the correct PIN, "Access Granted" is printed, correct becomes True, and the loop stops using break.  

#while:

correct_pin = 1234

correct = False
tries = 3
attempt = 0

while attempt < tries:
    entered_pin = int(input("Enter PIN: "))

    if entered_pin == correct_pin:
        print("Access Granted")
        correct = True
        break
    else:
        print("Wrong PIN,", tries - attempt - 1, "tries left")

    attempt += 1

if correct == False:
    print("Card Locked")

#tries = 3 stores the maximum number of attempts.attempt = 0 starts the counter.
#while attempt < tries runs until all attempts are used.
#The user enters a PIN using input().
#attempt += 1 moves to the next attempt.
#If all attempts are incorrect, print "Card Locked".
```

```python
#2 Grocery Checkout Total Your mom asked you to build a quick checkout tool for her small shop. 
# Items are entered one at a time and added to a running total. If the total crosses ₹1000 at any point,
#  a 10% discount should automatically apply to the rest of the checkout.
#  Given a list item_prices, write a program that loops through the prices, adding each to a running total. 
# Once the total crosses ₹1000, apply a 10% discount to all further items added. Print the final total.

# for loop :
item_prices = [300, 400, 500, 200]

total = 0
discount = False

for price in item_prices:
    if discount:
        total = total + (price * 0.9)
    else:
        total = total + price

    if total > 1000:
        discount = True

print("Final Total =", total)
#Add each item's price.
#Once total becomes more than 1000, apply a 10% discount to future items.

# while loop : 
item_prices = [300, 400, 500, 200]

total = 0
discount = False
i = 0

while i < len(item_prices):
    if discount:
        total = total + (item_prices[i] * 0.9)
    else:
        total = total + item_prices[i]

    if total > 1000:
        discount = True

    i += 1

print("Final Total =", total)
#i = 0 starts with the first item.while i < len(item_prices) continues until all items are processed.
#If discount is False, add the full price.If discount is True, add only 90% of the item's price (10% discount).
#Once the total exceeds ₹1000, set discount = True.i += 1 moves to the next item.
```

```python
# 3 Number Guessing Game Your little brother wants a guessing game where the computer picks a number and he keeps guessing until he gets it right. 
# After each guess, the program should print "Too High", "Too Low", or "Correct".
#  Given a secret_number and a list guesses (in the order they were made), write a program that loops through the guesses, printing "Too High" or "Too Low" for each using if-else,
#  and stops as soon as it finds the correct guess, printing "Correct" along with how many guesses it took.
 # for loop:

secret_number = 18
guesses = [5, 22, 15, 18]

count = 0

for guess in guesses:
    count += 1

    if guess > secret_number:
        print("Too High")
    elif guess < secret_number:
        print("Too Low")
    else:
        print("Correct")
        print("Guesses =", count)
        break

# Compare each guess.Print Too High or Too Low.
#Stop when the correct number is found.

# while loop : 

secret_number = 18
guesses = [5, 22, 15, 18]

count = 0
i = 0

while i < len(guesses):

    count += 1

    if guesses[i] > secret_number:
        print("Too High")
    elif guesses[i] < secret_number:
        print("Too Low")
    else:
        print("Correct")
        print("Guesses =", count)
        break

    i += 1

#i = 0 starts with the first guess.The loop continues while there are guesses left.
#Compare the current guess with the secret number.
#If it is greater, print Too High.If it is smaller, print Too Low.
#If it matches, print Correct and the number of guesses taken, then stop.
#i += 1 checks the next guess.
```

```python
# 4.You're helping at a local cinema's booking counter. 
# Tickets are sold one at a time until either all seats are sold or someone tries to book more seats than are available (in which case that request is rejected and booking stops). 
# Given total_seats and a list booking_requests (seats requested per booking, in order), write a program that loops through the requests, reducing the available seats each time. 
# If a request asks for more seats than are available, print "Not enough seats" and stop the loop. 
# Print the number of seats remaining at the end.
# for loop : 
total_seats = 20
booking_requests = [4, 5, 8, 6]

for seats in booking_requests:

    if seats <= total_seats:
        total_seats = total_seats - seats
    else:
        print("Not enough seats")
        break

print("Seats Remaining =", total_seats)
#Check each booking.If enough seats are available, subtract them.
#Otherwise stop booking.

# while loop : 
total_seats = 20
booking_requests = [4, 5, 8, 6]

i = 0

while i < len(booking_requests):

    if booking_requests[i] <= total_seats:
        total_seats = total_seats - booking_requests[i]
    else:
        print("Not enough seats")
        break

    i += 1

print("Seats Remaining =", total_seats)
#i = 0 starts with the first booking request.
#The loop runs until all requests are checked.
#If enough seats are available, subtract the requested seats.If the request is larger than the remaining seats, print Not enough seats and stop.
#i += 1 moves to the next booking.
#At the end, print the remaining seats.
```

```python
# 5.Study Streak Tracker You're building a habit tracker for your study group. 
# Each day's quiz score is checked one at a time. The streak continues as long as the score is 60 or above; the moment a score drops below 60, the streak breaks and processing stops. 
# Given a list scores (daily quiz scores in order), write a program that loops through the scores, using if-else to check each one: if it's 60 or above, increase the streak count; 
# if it drops below 60, print "Streak broken" and stop the loop. 
# Print the final streak length. i want all to be answered in for loop and while loop with easy methods and easy explanations
# fro loop : 
scores = [80, 75, 65, 58, 90]

streak = 0

for score in scores:

    if score >= 60:
        streak += 1
    else:
        print("Streak broken")
        break

print("Final Streak =", streak)
#If score is 60 or more, continue the streak.If score is below 60, stop the loop.
#Print the final streak count.

# while loop : 
scores = [80, 75, 65, 58, 90]

streak = 0
i = 0

while i < len(scores):

    if scores[i] >= 60:
        streak += 1
    else:
        print("Streak broken")
        break

    i += 1

print("Final Streak =", streak)
# i = 0 starts with the first score.
#The loop checks each score one by one.
# If the score is 60 or above, increase the streak count.If the score is below 60, print Streak broken and stop the loop.
#i += 1 moves to the next score.
#Finally, print the total number of consecutive days in the streak.
```
