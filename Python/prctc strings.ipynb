```python
#string : collection (sequence) of characters enclosed inside single quotes (' ') or double quotes (" ").
name = "Jayanth"
print(name)
#name is a variable."Jayanth" is stored as a string.
#print(name) displays the value.
username = "admin"
print("Welcome", username)
username = "admin"
#"Welcome" is a string.
#"admin" is another string.
```

```python
#printing a string : print() function to display strings.
name = "Joey"
print(name)
#multiple :
name = "Joey"
food = "Pizza"
print(name, "likes", food)
#Python automatically places a space between the arguments inside print().
```

```python
# string basics : A string is simply text.Python treats every character separately.
name = "Friends"
#Every character has an index number.
```

```python
# string indexes : An index is the position of each character in a string. counts from 0 
name = "Friends"

print(name[0])
print(name[3])
print(name[-1])
#name[0] → first character
#name[3] → fourth character
#name[-1] → last character
```

```python
# indexing : accessing one character using its position.
movie = "Friends"
print(movie[2])
# index 2 contains letter i 
movie = "Friends"
print(movie[-2])
```

```python
# string slicing : used to extract part of a string.
text = "I am Joey from Friends"

print(text[5:9])
#Starts at index 5 and stops before 9.
print(text[:4])
print(text[10:])
```

```python
# lower limit,upper limit,step size : 
#string[start:stop:step]
#Lower Limit. The starting position.

#Upper Limit.The ending position (not included).

#Step Size.How many characters to move each time.
text = "Python"

print(text[0:6:2])
```

```python
# reversing string : 
text = 'python'
print(text[::-1])
```

```python
# traversing string : visiting each character one by one.
name = "Python"

for ch in name:
    print(ch)
# The loop stores each character in ch and prints it.
# second char :
name = "Python"

for ch in name[::2]:
    print(ch)
```

```python
# string properties : Strings are immutable.This means you cannot change individual characters after creating a string.
name = "Python"

name = "Jython"

print(name)
```

```python
# concatenation : joining two or more strings together.
first = "Joey"
last = "Tribbiani"
print(first + " " + last)
# The + operator joins the strings.
city = "New "
country = "York"
print(city + country)
```
