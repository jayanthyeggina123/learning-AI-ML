```python
# for loop ; it is used to repeat the block of code for each item in a sequence such as list, string, or range
fruits = ["Apple", "Banana", "Mango"]

for fruit in fruits:
    print(fruit)

#fruits is a list.
#fruit stores one item from the list at a time.
#The loop prints each fruit one by one.
```

```python
#reversing a string : it can be reversed using slicing or a loop.
name = "Python"

print(name[::-1])
#[::-1] means:
#Start from the end.
#Move backwards one character at a time.
#The string is printed in reverse order.
```

```python
# range : this function generates a sequence of numbers. It is commonly used with a for loop.
for i in range(1, 6):
    print(i)
#Starts from 1.Stops before 6.
#Prints numbers from 1 to 5.
```

```python
# char :its not a separate data type in Python. A character is simply a string of length 1.
letter = "A"
print(letter)
#"A" is a string containing one character.
name = "Python"
for char in name:
    print(char)
```

```python
# while loop : repeats a block of code as long as the condition is True.
# Unlike a for loop, a while loop does not have an iterable variable. You usually control it using a counter variable.
pin = ""

while pin != "1234":
    pin = input("Enter ATM PIN: ")

print("Transaction Allowed")
#The user keeps entering the PIN.The loop continues until the correct PIN is entered.
```

```python
# Keep Asking Until User Types "yes"
answer = ""

while answer != "yes":
    answer = input("Do you want to continue? (yes/no): ")

print("Program Started")
```

```python
# counter variable : it keeps track of how many times a loop has executed.
count = 1

while count <= 3:
    print("Hello")
    count += 1
#count controls the loop.Each iteration increases count by 1.
#The loop executes 3 times.
```

```python
# break : break statement immediately stops the loop, even if the loop condition is still true.
for i in range(1, 11):
    if i == 5:
        break
    print(i)
# The loop starts printing numbers.# When i becomes 5, break stops the loop.
# Numbers after 4 are not printed.

students = ["Ram", "John", "Alice", "David"]
for student in students:
    if student == "Alice":
        print("Student Found")
        break
```

```python
# continue : this statement skips the current iteration and moves to the next iteration of the loop.
for i in range(1, 6):
    if i == 3:
        continue
    print(i)
    
#The loop prints numbers from 1 to 5.
#When i is 3, continue skips printing it.
#The loop continues with 4 and 5.
```
